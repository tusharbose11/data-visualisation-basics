```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10,8
import seaborn as sb
```

```python
#generating data for scatterplot and boxplot
no_of_skills = np.linspace(1,30, 500, dtype=int)
buffer = np.random.randint(1, 40000, 500)
buffer = buffer
salaries = (no_of_skills * 8000) + buffer
no_of_skills[:5], buffer[:5], salaries[:5]
udf = pd.DataFrame({'no_of_skills':no_of_skills, 'salaries' : salaries})
```

### 2.1) Pie Chart

```python
#pie chart
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
#Tushar Bose - 189302021
```

### 2.2) Bar Chart

```python
#bar chart
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
xpos = np.arange(len(company))
xpos
plt.bar(xpos,revenue, label="Revenue")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
#Tushar Bose - 189302021
```

### 2.3) Box Plot

```python
#boxplots
udf.plot(kind='box', subplots=True, layout=(2,2), sharex=False, sharey=False)
#Tushar Bose - 189302021
```

### 2.5) Line Graphs

```python
# line graph
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]

plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')


plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
#Tushar Bose - 189302021
```
